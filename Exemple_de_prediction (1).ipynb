```python
# prompt: donne un autre exemple  de code prediction sur ce jeux de donnees

import yfinance as yf
from sklearn.linear_model import LinearRegression
import pandas as pd

# Define the ticker symbol
ticker_symbol = 'AAPL'

# Download historical data
aapl_data = yf.Ticker(ticker_symbol)
aapl_history = aapl_data.history(period="2y")  # Increased period for better prediction

# Prepare the data for prediction
# Use 'Close' price as the target variable
data = aapl_history[['Close']]
data['Tomorrow_Close'] = data['Close'].shift(-1)  # Create a new column for the next day's closing price
data = data.dropna()  # Drop rows with missing values

# Feature Engineering (example: using rolling averages)
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data = data.dropna()

# Split the data into training and testing sets
train_size = int(len(data) * 0.8)
train_data, test_data = data[:train_size], data[train_size:]

# Define the features (X) and target (y)
X_train = train_data[['SMA_5', 'SMA_20']]
y_train = train_data['Tomorrow_Close']
X_test = test_data[['SMA_5', 'SMA_20']]
y_test = test_data['Tomorrow_Close']

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model (example: using Mean Absolute Error)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae}")


# Example prediction for the next day (using the last available data point)
last_data_point = data.tail(1)[['SMA_5', 'SMA_20']]
next_day_prediction = model.predict(last_data_point)
print(f"Predicted closing price for the next day: {next_day_prediction[0]}")

# Important: This is a very simple prediction model.
# In real-world scenarios, you'd need more sophisticated features and models.
# Consider using more advanced techniques, time series analysis, and proper backtesting.
# This example is for illustrative purposes only and should not be used for actual trading.
```
